## The Numerical Derivative


```python
def f(x):
  """
  定义一个简单的二次函数。
  """
  return 2 * x**2


x = np.arange(0, 5, 0.001)
y = f(x)
plt.plot(x, y)

# 定义一个非常小的增量，用于近似计算导数
p2_delta = 0.0001
x1 = 2
x2 = x1 + p2_delta
y1 = f(x1)
y2 = f(x2)
print((x1, y1), (x2, y2))

# 切线的近似导数
approximate_derivative = (y2 - y1) / (x2 - x1)
# 和截距(y-intercept)
b = y2 - approximate_derivative * x2


# 定义切线函数
def tangent_line(x):
  """
  计算切线的 y 值。

  # Returns:
  切线在 x 处的 y 值
  """
  return approximate_derivative * x + b


# 选择要绘制切线的 x 值范围
to_plot = [x1 - 0.9, x1, x1 + 1.9]
# 绘制切线
plt.plot(to_plot, [tangent_line(i) for i in to_plot])

print("f(x)的近似导数为：", f"x = {x1} 是 {approximate_derivative}")
plt.show()
```

生成多条切线


```python
plt.plot(x, y)
colors = ["k", "g", "r", "b", "c"]


def approximate_tangent_line(x, approximate_derivative):
  return approximate_derivative * x + b


for i in range(5):
  p2_delta = 0.0001
  x1 = i
  x2 = x1 + p2_delta

  y1 = f(x1)
  y2 = f(x2)

  print((x1, y1), (x2, y2))
  approximate_derivative = (y2 - y1) / (x2 - x1)
  b = y2 - (approximate_derivative * x2)

  to_plot = [x1 - 0.9, x1, x1 + 0.9]

  plt.scatter(x1, y1, c=colors[i])
  plt.plot(
    [point for point in to_plot],
    [approximate_tangent_line(point, approximate_derivative) for point in to_plot],
    c=colors[i],
  )

  print("Approximate derivative for f(x)", f"where x = {x1} is {approximate_derivative}")

plt.show()
```
